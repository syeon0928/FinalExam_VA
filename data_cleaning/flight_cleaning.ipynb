```python
airport_distance = pd.merge(distance,airport_code, left_on='ORIGIN', right_on='IATA', how='left')
```

```python
airport_distance = airport_distance.dropna(subset=["Name"])
irport_distance = airport_distance.drop(columns=['ORIGIN_AIRPORT_SEQ_ID', 'DEST_AIRPORT_SEQ_ID','ICAO','Latitude','Longitude'])
airport_distance.rename(columns = {'Name':'AIRPORT NAME', 'City':'CITY', 'Country':'COUNTRY'}, inplace = True)
```
